```python
# Importamos librer√≠as necesarias
import pandas as pd
import os
import glob
import seaborn as sns
```

```python
csv_files = glob.glob('*.csv')
lista_data = []
  
# Creamos un bucle que recorre cada archivo csv de la carpeta, crea un df y lo agrega a una lista
for filename in csv_files:
    data = pd.read_csv(filename)
    lista_data.append(data)
```

```python
# Diccionario que contiene cada df y le asigna como key el nombre del archivo correspondiente
dicc_df = {'State_Zhvi_1Bedroom':lista_data[0], 'State_Zhvi_2Bedroom':lista_data[1],
           'State_Zhvi_3Bedroom':lista_data[2], 'State_Zhvi_4Bedroom':lista_data[3],
           'State_Zhvi_5Bedroom':lista_data[4], 'State_Zhvi_AllHomes':lista_data[5],
           'State_Zhvi_Condominum':lista_data[6], 'State_Zhvi_SingleFamilyResidence':lista_data[7]}
```

```python
for key in dicc_df:
    print(key,'\n\n',dicc_df[key].head(5),'\n\n')
```

```python
for key in dicc_df:
    print(key,'\n\n',dicc_df[key].info(),'\n\n')
```

```python
for key in dicc_df:
    print(key,'\n\n',dicc_df[key].describe(exclude='object'),'\n\n')
```

## - Valores Nulos

```python
# Buscamos registros nulos por columna
for key in dicc_df:
    col = dicc_df[key].columns
    print(key,'\n')
    for e in col:
        print(e,'\t',round(len(dicc_df[key][e][dicc_df[key][e].isnull()])/len(dicc_df[key])*100,2),'%')
    print('\n\n')
```

```python

for key in dicc_df:
    print(key,':',dicc_df[key].duplicated().sum())
```

## -- Outliers

```python
def det_outlier_df_vf(dicc_df):

    for key in dicc_df:
        
        for columna in dicc_df[key].columns:
            
            if dicc_df[key].loc[:,columna].dtype!=object:

                Q1 = dicc_df[key][columna].quantile(0.25)
                Q3 = dicc_df[key][columna].quantile(0.75)
                IQR = Q3 - Q1
                BI = Q1 - 1.5*IQR
                BS = Q3 + 1.5*IQR

                out_sup = dicc_df[key][dicc_df[key][columna] > BS].index
                out_inf = dicc_df[key][dicc_df[key][columna] < BI].index
                outliers = []
                for i in out_sup:
                    outliers.append(i)
                    for j in out_inf:
                        outliers.append(j)
                size = len(outliers)


                print('\n')
                print(key)
                print(columna)
                print('Q1: ',Q1)
                print('Q3: ',Q3)
                print('IQR: ',IQR)
                print('BI: ',BI)
                print('BS: ',BS)
                print (f'Hay {size} valores at√≠picos en la variable {columna}, y corresponden a los √≠ndices: {outliers}')
```

```python
det_outlier_df_vf(dicc_df)
```
