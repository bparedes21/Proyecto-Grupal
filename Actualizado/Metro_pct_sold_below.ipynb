```python
#esta funcion recibe como parametro un df
#combierte las columnas anio en registros dentro de la columna Date
#los registros precio dentro  de la columna Price
#devuelve un dataframe con las columnas ["RegionName","State","Metro","CountyName","SizeRank","Date",	"Price"]
def func_crear_col_Date_price(df):
    
    #realizo una copia del dataframe. 
    import pandas as pd
    df_col_anio=df.copy()

    #lista de columnas distintas a anio 
    list_columnas_validas=['Unnamed: 0','RegionID', 'SizeRank', 'RegionName', 'RegionType', 'StateName']

    #en este dataframe solo quedan las columnas año
    df_col_anio=df_col_anio.drop(columns=list_columnas_validas)

    #hago una lista con las columnas anio
    list_columns=df_col_anio.columns

    #creo un nuevo df con las columnas distintas a anio
    df2=df[['Unnamed: 0','RegionID', 'SizeRank', 'RegionName', 'RegionType', 'StateName']]
    
    #lista sin la columna Unnamed: 0 ni anios
    list_columnas_df2=['RegionID', 'SizeRank', 'RegionName', 'RegionType', 'StateName']
    
    #para recorrer el df tomo la columna unnamed 0 como indice guardo en la variable tamanio_df el valor maximo y sumo 1
    tamanio_df=df["Unnamed: 0"].max()
    
    contador=0
    #creo un dataframe nuevo para concatenar los dataframes procesados
    new_df=pd.DataFrame(columns=['RegionID', 'SizeRank', 'RegionName', 'RegionType', 'StateName','Date','Price'])
    #lista de las que voy a agregar en df1
    list_columnas_df2=['RegionID', 'SizeRank', 'RegionName', 'RegionType', 'StateName']
    list_coluname=df["Unnamed: 0"].unique()
    for i, dato in reversed(list(enumerate(list_coluname))):
        #print(dato)
        #if(dato!=tamanio_df):
            #new_df=pd.read_csv("new_fd.csv")
            #print(new_df)
        serie_col_val=df2.loc[dato]
        registros_anio=df_col_anio.iloc[dato]
        df1=pd.DataFrame(registros_anio)
        df1=df1.reset_index()
        df1=df1.rename(columns={"index":"Date",dato:"Price"})
        list_columnas_df2=['RegionID', 'SizeRank', 'RegionName', 'RegionType', 'StateName']
        
        #df1.to_csv("df1.csv",index=False)
        #df1=pd.read_csv("df1.csv")
        for e in list_columnas_df2:
            var_aux=serie_col_val[e]
            df1.insert(contador, e , var_aux, allow_duplicates=False)
            
        #cambio el orden de las columnas        
        new_df = pd.concat([df1, new_df], axis=0) 
        #new_df.to_csv("new_fd.csv",index=False)
    #new_df=new_df[["RegionName","State","Metro","CountyName","SizeRank","Date","Price"]] 
    
    
        #uno los df
    
    return new_df
```
